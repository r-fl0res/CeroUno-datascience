```python
#6ta pregunta  aqui quedo asi porque al principio no sabia que solo serian las unicas, por lo que 
#habia impreso todas las palabras y las veces que se repetian. por eso quedan asi
def mayus(y):
    preg1=leer(y)
    mayus = str.upper(preg1)
    listmayus = mayus.split()
    setmayus =set(listmayus)
    unimayus = list(setmayus)

    for i in unimayus:
        cadena1 =str(unimayus.count(i))
        print(i+':'+cadena1)
```

```python
#8pregunta
def spmarks(y):
    preg1= leer(y)
    letras= 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ '
    letrasl=list(letras)
    texto = list(preg1)
    sl1 = set(letrasl)
    stxt = set(texto)
    diff = stxt.difference(sl1)
    diff.remove('\n')
    print(diff)
```

```python
spmarks("8800.txt")
```
